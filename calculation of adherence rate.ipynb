```python
import pandas as pd
from random import randint
from statistics import mode
import random
random.seed(0)
```

```python
df_pipo = pd.read_excel("PIPO.xlsx")
df_invoice = pd.read_excel("invoice data.xlsx")
```

#### making a list of unique clients

```python
# make a list of unique clients
client_unique = []  # [1018,1019,1020,1021,1023,1024,...] 870 unique clients
for i in range(len(df_invoice)):
    if df_invoice["#Rno"][i] not in client_unique:
        client_unique.append(df_invoice["#Rno"][i])
```

```python
len(client_unique)
```

#### making a dictionary having keys as Rno and values as their workout AR rate

```python
client_actualworkout_count_dict = {}  
```

```python
for client in client_unique:
    counter = 0
    for i in range(len(df_pipo)):
        if df_pipo["#Rno"][i] == client:
            counter += 1
    client_actualworkout_count_dict[client] = counter
```

```python
client_actualworkout_count_dict
```

#### making a dict with keys as Rno and values as their total invoice validity period in days

```python
client_totalDuration_dict = {}
```

```python
for c in client_unique:
    client_allRecharges_duration_list = []
    for i in range(len(df_invoice)):
        if df_invoice["#Rno"][i] == c:
            client_allRecharges_duration_list.append(df_invoice["package"][i])
            
    client_totalDuration_dict[c] = (sum(client_allRecharges_duration_list))*30  # in days           
```

```python
client_totalDuration_dict
```

#### using both the list having AR and total periods to create an another dict having  % of duration a client worksout in his/her validy period in the gym as values

```python
client_dedication = {}  # key : Rno and value : duration (in %) of time a clients workouts in the gym
```

```python
for key1,actual_duration in client_actualworkout_count_dict.items():
    for  key2,total_duration in client_totalDuration_dict.items():
        if key1 == key2:
            client_dedication[key1] = round((actual_duration/total_duration)*100,3)
        
```

```python
len(client_dedication)
```

```python
client_dedication
```

```python
df_AR_of_all494_clients = pd.DataFrame(list(client_dedication.items()),columns = ['#Rno','AR']) 
df_AR_of_all494_clients.to_excel("all clients AR.xlsx", index = False)
```

#### if AR is above 75 then a client is a High AR client

```python
tolerance = 75
high_AR_clients = [] # list of high Ar clients
for k,v in client_dedication.items():
    if v>tolerance:
        high_AR_clients.append(k)

len(high_AR_clients)
```

```python
df_35_highAR_clients  = pd.DataFrame(high_AR_clients, columns = ["#Rno"])
df_35_highAR_clients.head()
```

```python
df_35_highAR_clients.to_excel("#Rno's of 35 high AR clients.xlsx",index = False)
```
